W przypadku obrazów w odcieniach szarości pojedynczy piksel z zakresu [0; 255] reprezentowany jest jako 8-bitowa liczba bez znaku.
Pewnym rozszerzeniem analizy sposobu reprezentacji obrazu może być następujący eksperyment.
Załóżmy, że z każdego z 8 bitów możemy stworzyć pojedynczy obraz binarny (ang. _bit-plane slicing_).
Dla obrazka _100zloty.jpg_ (https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/100zloty.jpg) stwórz 8 obrazów, z których każdy powinien zawierać jedną płaszczyznę bitową.
Podpowiedź $-$ warto sprawdzić, jak realizuje się bitowe operacje logiczne.
Zastosowanie takiej dekompozycji obrazu pozwala na analizę ,,ważności'' poszczególnych bitów.
Jest to użyteczne w kwantyzacji, ale także w kompresji.

W drugim etapie zadania proszę spróbować odtworzyć obraz oryginalny z mniejszej liczby obrazów binarnych.
Warto zacząć od dwóch najbardziej znaczących bitów, a później dodawać kolejne.
Należy utworzyć co najmniej trzy wersje zrekonstruowanych obrazów.
Podpowiedź $-$ rekonstrukcja obrazu to mnożenie przez odpowiednią potęgę liczby 2 (przesunięcie bitowe) oraz dodawanie.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os


if not os.path.exists("100zloty.jpg"):
  !wget https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/100zloty.jpg

I = cv2.imread("100zloty.jpg")
Ig = cv2.cvtColor(I, cv2.COLOR_RGB2GRAY)


plt.imshow(I)
plt.xticks([]), plt.yticks([])
plt.title("100zł - colour")
plt.show()

plt.imshow(Ig)
plt.gray()
plt.xticks([]), plt.yticks([])
plt.title("grey 100zł")
plt.show()
```

```python
bin_pic = []
rows = Ig.shape[0]
cols = Ig.shape[1]
eight=[] 
seven=[]
six=[]
five=[]
four=[]
three=[]
two=[]
one=[]

layers = [eight, seven, six, five, four, three, two, one]

for i in range(rows):
  for j in range(cols):
    bin_pic.append(np.binary_repr(Ig[i][j] ,width=8)) # zamienia każdy bit na zapis binarny w stringu, width definiuje długość zapisu binarnego, my chcemy 8, żeby mieć cały bajt

#Dostajemy pełny obraz zapisany w stringach reprezentujących binarną wartość. Stringi pozwalają na indeksowanie, więc teraz w pętlach iteracyjnie "wydłubujemy" kolejne warstwy bitów.

for bajt in bin_pic:
  # for num in layers:
  #   i=0
  #   num.append(int(bajt[i]))
  #   i+=1
  #nie działało :<
  eight.append(int(bajt[0]))
  seven.append(int(bajt[1]))
  six.append(int(bajt[2]))
  five.append(int(bajt[3]))
  four.append(int(bajt[4]))
  three.append(int(bajt[5]))
  two.append(int(bajt[6]))
  one.append(int(bajt[7]))


#Zapisanie listy z powrotem jako macierz
#nie mam pojęcia czemu, ale list comprehension, ani nic z wywoływaniem kolejnych elementów layers nie działało
eight = np.array(eight)
eight = eight.reshape(rows,cols)
seven = np.array(seven)
seven = seven.reshape(rows,cols)
six = np.array(six)
six = six.reshape(rows,cols)
five = np.array(five)
five = five.reshape(rows,cols)
four = np.array(four)
four = four.reshape(rows,cols)
three = np.array(three)
three = three.reshape(rows,cols)
two = np.array(two)
two = two.reshape(rows,cols)
one = np.array(one)
one = one.reshape(rows,cols)


plt.imshow(eight)
plt.xticks([]), plt.yticks([])
plt.title("eighth bit")
plt.show()

plt.imshow(seven)
plt.xticks([]), plt.yticks([])
plt.title("seventh bit")
plt.show()

plt.imshow(six)
plt.xticks([]), plt.yticks([])
plt.title("sixth bit")
plt.show()

plt.imshow(five)
plt.xticks([]), plt.yticks([])
plt.title("fifth bit")
plt.show()

plt.imshow(four)
plt.xticks([]), plt.yticks([])
plt.title("fourth bit")
plt.show()

plt.imshow(three)
plt.xticks([]), plt.yticks([])
plt.title("third bit")
plt.show()

plt.imshow(two)
plt.xticks([]), plt.yticks([])
plt.title("second bit")
plt.show()

plt.imshow(one)
plt.xticks([]), plt.yticks([])
plt.title("first bit")
plt.show()
```

```python
eight_rec = eight * 2**7
seven_rec = seven * 2**6
six_rec = six * 2**5
five_rec = five * 2**4
four_rec = four * 2**3
three_rec = three * 2**2
two_rec = two * 2

plt.imshow(Ig)
plt.xticks([]), plt.yticks([])
plt.title("Original grey 100zł")
plt.show()

plt.imshow(seven_rec + eight_rec + six_rec)
plt.xticks([]), plt.yticks([])
plt.title("eighth, seventh and sixth bit pic")
plt.show()

plt.imshow(seven_rec + eight_rec)
plt.xticks([]), plt.yticks([])
plt.title("eighth and seventh bit pic")
plt.show()

plt.imshow(six_rec + seven_rec)
plt.xticks([]), plt.yticks([])
plt.title("seventh and sixth bit picture")
plt.show()

#Kolejne obrazy bitowe dodają więcej detali - najwięcej informacji jest zapisanych w najstarszych bitach, więc jeżeli chcemy zaoszczędzić trochę miejsca, możemy pozbyć się młodszych bitów
#nie tracąc za bardzo na jakości obrazu
```
